# Chapter 7

```python
# Load libraries
import pandas as pd
import numpy as np
```

```python
# Load the datafile
candy_df = pd.read_excel(r'Data/candyhierarchy2017.xlsx')

# Validate the dataload
candy_df.head()
```

```python
# Obtain the shape of the dataset
candy_df.shape
```

## Filter out missing data

```python
# Obtain the NA counts by column
nacounts = candy_df.isna().sum()

# Sort & display the top 10 nacounts
nacounts.sort_values(ascending=False).head(10)
```

```python
# The "Unnamed: 113" column is only populated in 9 rows. This column will be removed from the dataset.
candy_df = candy_df.drop(['Unnamed: 113'], axis=1)
```

```python
# There are multiple incomplete surveys in the dataset. Drop surveys that contain more than 25% blank values.
candy_df = candy_df.dropna(thresh=30)
```

```python
# View the shape of our filtered dataframe
candy_df.shape
```

## Fill in missing data

```python
# Missing data in the Q12 fields will be replaced with 0 as this is a binary field
candy_df["Q12: MEDIA [Daily Dish]"] = candy_df["Q12: MEDIA [Daily Dish]"].fillna(0)
candy_df["Q12: MEDIA [Science]"] = candy_df["Q12: MEDIA [Science]"].fillna(0)
candy_df["Q12: MEDIA [ESPN]"] = candy_df["Q12: MEDIA [ESPN]"].fillna(0)
candy_df["Q12: MEDIA [Yahoo]"] = candy_df["Q12: MEDIA [Yahoo]"].fillna(0)
```

```python
# Obtain a list of column names that excludes Q7, Q8, & Q9 as these are manual other fields that we do not want to fill NAs on.
columns = list(candy_df.columns)
unwanted = ['Q9: OTHER COMMENTS', 'Q8: DESPAIR OTHER', 'Q7: JOY OTHER']
for element in unwanted:
    columns.remove(element)
    
columns
```

```python
# Fill the na values from the columns specified in the previous step
candy_df[columns] = candy_df[columns].fillna(method='ffill')
```

# Chapter 8

## Create hierarchial index

```python
# Use set index to create the hierarchial index
hier_candy_df = candy_df.set_index(['Q1: GOING OUT?', 'Q2: GENDER', 'Internal ID'])

# View the dataframe
hier_candy_df
```

## Combine and Merge Datasets

```python
# Create a database for demographics
candy_demo_df = candy_df.iloc[:,:6]

# Drop Q1
candy_demo_df = candy_demo_df.drop(['Q1: GOING OUT?'], axis=1)

# View the dataframe
candy_demo_df
```

```python
# Create a database for responses
candy_resp_df = candy_df.drop(['Q2: GENDER', 'Q3: AGE', 'Q4: COUNTRY', 'Q5: STATE, PROVINCE, COUNTY, ETC'], axis=1)

# View the dataframe
candy_resp_df
```

```python
# Merge the dataframes
pd.merge(candy_demo_df, candy_resp_df, on='Internal ID')
```

# Chapter 10

## Grouping with Index Levels

```python
# Get counts of responses by gender
hier_candy_df.groupby(level='Q2: GENDER', axis=0).count()
```

## Split/Apply/Combine

```python
# Define a function to obtain the 5 youngest records
def oldest(df, n=5, column='Q3: AGE'):
    return df.sort_values(by=column)[:n]

# Replace any strings in the Age column with NAN
candy_df['Q3: AGE'] = candy_df['Q3: AGE'].apply(pd.to_numeric,errors='coerce')

# Apply the function after grouping by gender
candy_df.groupby('Q2: GENDER').apply(oldest)
```

```python
# Load the datafile
boing_df = pd.read_excel(r'Data/BOING-BOING-CANDY-HIERARCHY-2016-SURVEY-Responses.xlsx')

# View the dataframe
boing_df
```

## Convert between string and date time

```python
# Convert to string
boing_df['string time'] = boing_df['Timestamp'].astype(str)

# View the new column
boing_df['string time']
```

```python
# Convert string to time
boing_df['Time'] = pd.to_datetime(boing_df['string time'])

# View the new column
boing_df['Time']
```

## Convert timestamps to periods and back

```python
# Set Timestamp as the index
boing_df = boing_df.set_index(['Timestamp'])

# Convert to periods with a daily frequency
period_df = boing_df.to_period(freq='D')
# view the dataframe
period_df
```

```python
# Convert from daily period to timestamp at end of day
ts_df = period_df.to_timestamp(how='end')

# View the dataframe
ts_df
```
