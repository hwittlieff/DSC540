# Project Milestone 5: Merging the Data and Storing in a Database/Visualizing Data

```python
# Load libraries
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Establish connection
conn = sqlite3.connect('mydata.sqlite')
```

Load the Netflix_Kaggle table (Note: this table was added to the SQL database in Milestone 2)

```python
# Netflix Kaggle query
kaggle_query = pd.read_sql_query ('''Select * From Netflix_Kaggle ''', conn)
```

```python
# Load kaggle table
kaggle_df = pd.DataFrame(kaggle_query, columns = ['Title', 'Director', 'Cast', 'Country', 'Release_Year',
                                                 'Rating', 'Duration', 'Description', 'Year_Added', 'AGE', 'Title_Year'])
```

```python
# Validate the data loaded correctly
kaggle_df.head(20)
```

Load the Netflix_Wiki table (Note: this table was added to the SQL database in Milestone 3)

```python
# Netflix Wiki query
wiki_query = pd.read_sql_query ('''Select * From Netflix_WIki ''', conn)
```

```python
# Load wiki table
wiki_df = pd.DataFrame(wiki_query, columns = ['Year', 'Revenue_in_Millions', 'Net_Income_in_Millions', 'Price_per_share',
                                                 'Employees', 'Paid_Memberships_in_Millions', 'Fortune_500_Rank'])
```

```python
# Validate the data loaded correctly
wiki_df.head(16)
```

Load the Netflix_OMDB table (Note: this table was added to the SQL database in Milestone 4)

```python
# Netflix OMDB query
OMDB_query = pd.read_sql_query ('''Select * From Netflix_OMDB ''', conn)
```

```python
# Load OMDB table
OMDB_df = pd.DataFrame(OMDB_query, columns = ['Title', 'Year', 'Rated', 'Released', 'Runtime', 'Genre', 'Director',
                                             'Writer', 'Actors', 'Plot', 'Language', 'Country', 'Awards', 'Metascore', 
                                             'IMDBRating', 'IMDBVotes', 'IMDBID', 'BOXOFFICE'])
```

```python
# Validate the data loaded correctly
OMDB_df.head()
```

```python
# Close the connection
conn.close()
```

Merge the Kaggle & OMDB dataframes

```python
# Join the two dataframes
kaggle_omdb_df = OMDB_df.join(kaggle_df.set_index('Title'), on='Title', lsuffix='_OMDB', rsuffix='_Kaggle')
```

```python
# Validate the merge
kaggle_omdb_df.head()
```

Merge with the wiki dataframe

```python
# Group the previously merged dataframe by year
k_o_annual_df = kaggle_omdb_df.groupby('Year_Added', as_index=False).mean()
```

```python
# Drop age & Year_Added fields
k_o_annual_df = k_o_annual_df.drop(['Release_Year', 'AGE'], axis = 1)
```

```python
# View the dataframe
k_o_annual_df.head()
```

```python
# Join the two dataframes
df = wiki_df.join(k_o_annual_df.set_index('Year_Added'), on='Year', lsuffix='_OMDB', rsuffix='_Kaggle')
```

```python
# Replace blanks with np.nan
df = df.replace(r'^\s*$', np.nan, regex=True)
```

```python
# View the dataframe
df.head(17)
```

## Visualization 1: Netflix Paid Memberships by Year

```python
# Create a bar chart of the paid memberships by year
plt.bar(df['Year'], df['Paid_Memberships_in_Millions'])
plt.xlabel('Year')
plt.ylabel('Paid Memberships (in millions)')
plt.locator_params(axis="x", nbins=4)
plt.show()
```

## Visualization 2: Price Per Share vs Paid_Memberships

```python
# Price Per Share vs Paid_Memberships
fig=plt.figure()
ax=fig.add_axes([0, 0, 1, 1])
ax.scatter(df['Paid_Memberships_in_Millions'], df['Price_per_share'], color='r')
ax.set_xlabel('Paid Memberhips (in millions)')
ax.set_ylabel('Price Per Share')
ax.set_title('Netflix Subscribers vs Share Price')
plt.show()
```

## Visualization 3: Average IMDBRating vs Paid Memberships

```python
# Average IMDB Rating vs memberships
fig=plt.figure()
ax=fig.add_axes([0, 0, 1, 1])
ax.scatter(df['IMDBRating'], df['Paid_Memberships_in_Millions'], color='b')
ax.set_xlabel('Average IMDBRating (Films added each year)')
ax.set_ylabel('Paid Memberhips (in millions)')
ax.set_title('Average IMDB Rating vs Paid Memberships')
plt.show()
```

## Visualization 4: Average Runtime vs Paid Memberships

```python
# Average Runtime vs memberships
fig=plt.figure()
ax=fig.add_axes([0, 0, 1, 1])
ax.scatter(df['Runtime'], df['Paid_Memberships_in_Millions'], color='g')
ax.set_xlabel('Average Runtime (Films added each year)')
ax.set_ylabel('Paid Memberhips (in millions)')
ax.set_title('Average Runtime vs Paid Memberships')
plt.show()
```

## Visualization 5: Average Metascore vs Paid Memberships

```python
# Average Runtime vs memberships
fig=plt.figure()
ax=fig.add_axes([0, 0, 1, 1])
ax.scatter(df['Metascore'], df['Paid_Memberships_in_Millions'], color='b')
ax.set_xlabel('Average Metascore (Films added each year)')
ax.set_ylabel('Paid Memberhips (in millions)')
ax.set_title('Average Metascore vs Paid Memberships')
plt.show()
```
