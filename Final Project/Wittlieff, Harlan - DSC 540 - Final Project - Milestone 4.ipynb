```python
# Assign the portal to a variable
base_url = 'http://www.omdbapi.com/?'
```

```python
# Establish empty lists for variables
title = []
year = []
rated = []
released = []
runtime = []
genre = []
director = []
writer = []
actors = []
plot = []
language = []
country = []
awards = []
metascore = []
imdbrating = []
imdbvotes = []
imdbid = []
boxoffice = []
f_type = []
```

## Data Transformation 1: Inserting np.nan for series shows box office revenue

Below is a function for appending the various lists of data with the data received from the API call. Since there is not field for "BoxOffice" for any of the titles that are a series show, np.nan will be inserted. This is done via a try/except block towards the bottom of the function

```python
# Define a function to add a movie's data to the list
def add_to_list(j_data):
    # Append the lists for the newly obtained data
    title.append(j_data['Title'])
    year.append(j_data['Year'])
    rated.append(j_data['Rated'])
    released.append(j_data['Released'])
    runtime.append(j_data['Runtime'])
    genre.append(j_data['Genre'])
    director.append(j_data['Director'])
    writer.append(j_data['Writer'])
    actors.append(j_data['Actors'])
    plot.append(j_data['Plot'])
    language.append(j_data['Language'])
    country.append(j_data['Country'])
    awards.append(j_data['Awards'])
    metascore.append(j_data['Metascore'])
    imdbrating.append(j_data['imdbRating'])
    imdbvotes.append(j_data['imdbVotes'])
    imdbid.append(j_data['imdbID'])
    f_type.append(j_data['Type'])
    
    # Series shows do not contain information for box office, NAN will be input in these instances
    try:
        boxoffice.append(j_data['BoxOffice'])
    except:
        boxoffice.append(np.nan)
```

```python
# Create the function to search for a movie
def find_movie(title):
    try:
        #Build the url from the base, title, & api key
        url = base_url + urllib.parse.urlencode({'t': str(title)})+apikey
        print(f'Retrieving the data of "{title}" now... ')
        print(url)
        uh = urllib.request.urlopen(url)
        data = uh.read()
        json_data=json.loads(data)
        
        # If a response is received, add the movie's data to the list
        if json_data['Response']=='True':
            add_to_list(json_data)
            print('Added to list...')
        # If a valid response is not received, print an error
        else:
            print("Error encountered: ",json_data['Error'])
    except urllib.error.URLError as e:
        print(f"ERROR: {e.reason}")
```

```python
# Load the list of movies
titles_df = pd.read_csv(r'Data/titles.csv')
```

```python
for i in titles_df['title']:
    find_movie(i)
```

## Data Transformation 2: Building a dataframe & adding headers

Now that the data is contained within individual lists, a dataframe will have to be created

```python
# Build Dataframe from lists
df = pd.DataFrame(list(zip(title, year, rated, released, runtime, genre, director, writer, actors,
                          plot, language, country, awards, metascore, imdbrating, imdbvotes, imdbid,
                          boxoffice, f_type)), columns=['Title', 'Year', 'Rated', 'Released', 'Runtime',
                                                       'Genre', 'Director', 'Writer', 'Actors', 'Plot',
                                                       'Language', 'Country', 'Awards', 'Metascore', 'IMDBRating',
                                                       'IMDBVotes', 'IMDBID', 'Box Office Revenue', 'Type'])
```

## Data Transformation 3: Adding NANs

The current data contains the string "N/A" instead of true NAN values. These need to be replaced.

```python
# Use the replace function to replace string "N/A" with np.nan
df = df.replace('N/A', np.nan)
```

```python
# Retreive nan counts by column
df.isna().sum()
```

## Data Transformation 4: Removing rows with NAN in the IMDBRating field

One of the major pieces of information needed from this dataset is the IMDB Rating. Due to this any title that is missing data on this field will be removed

```python
# Remove rows with nan in the IMDBRating column
df = df[df['IMDBRating'].notna()]
```

```python
# Validate the change by reviewing the nan counts
df.isna().sum()
```

## Data Transformation 5: Removing "min" from Runtime field

Removing the string "min" and converting the field to numeric will be beneficial for future analysis.

```python
# Use the replace function to remove the "min" string
df['Runtime'] = df['Runtime'].str.replace(' min', '')

# Convert to numeric
df['Runtime'] = pd.to_numeric(df['Runtime'], downcast='integer')
```
